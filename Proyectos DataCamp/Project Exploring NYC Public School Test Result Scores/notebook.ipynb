Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

## Load Data

# Which NYC schools have the best math results?

```python
# Calculate the best math results threshold (80% of 800)
best_math_threshold = 0.8 * 800
```

```python
# Filter the schools with math scores at least 80% of the maximum possible score
best_math_schools = schools[schools['average_math'] >= best_math_threshold]
best_math_schools
```

```python
# Select only the 'school_name' and 'average_math' columns
best_math_schools = best_math_schools[['school_name', 'average_math']]
best_math_schools
```

```python
# Sort the DataFrame by 'average_math' in descending order
best_math_schools = best_math_schools.sort_values('average_math', ascending=False)
best_math_schools
```

```python
# Now best_math_schools contains the schools with the best math results
print(best_math_schools)
```

# What are the top 10 performing schools based on the combined SAT scores?

```python
# Calculate the total SAT score for each school
schools['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']
schools.head()
```

```python
# Sort the schools by the total SAT score in descending order
schools_sorted = schools.sort_values('total_SAT', ascending=False)
schools_sorted
```

```python
# Select the top 10 schools
top_10_schools = schools_sorted.head(10)
top_10_schools
```

```python
# Create a new DataFrame with the top 10 schools and their total SAT scores
top_10_schools = top_10_schools[['school_name', 'total_SAT']]
top_10_schools
```

```python
# Now top_10_schools contains the top 10 schools with the highest total SAT scores
print(top_10_schools)
```

# Which single borough has the largest standard deviation in the combined SAT score?

```python
# Group by borough and calculate statistics
borough_stats = schools.groupby('borough').agg(
    num_schools=('school_name', 'count'),
    average_SAT=('total_SAT', 'mean'),
    std_SAT=('total_SAT', 'std')
).reset_index()

borough_stats
```

```python
# Find the borough with the largest standard deviation of total_SAT
largest_std_dev_borough = borough_stats.loc[borough_stats['std_SAT'].idxmax(), 'borough']
largest_std_dev_borough
```

```python
# Create a new DataFrame for the borough with the largest standard deviation
largest_std_dev = borough_stats[borough_stats['borough'] == largest_std_dev_borough]
largest_std_dev
```

```python
# Round all numeric values to two decimal places
largest_std_dev = largest_std_dev.round(2)
largest_std_dev
```

```python
# Now largest_std_dev contains the borough with the largest standard deviation of total_SAT
print(largest_std_dev)
```
